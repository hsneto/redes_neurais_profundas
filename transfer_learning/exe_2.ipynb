````python
from transferlearning_v2 import Hyperparameters, train, Extract
from params_analysis import save_params, load_df
import time
import matplotlib.pyplot as plt

%matplotlib inline  
````

````python
LOGDIR = 'results/exe_2/'
CPP = './tensorboard/cifar10_vgg16/exe_2/'

JSON_FILE = 'exe_2.json'
CUT_LAYER = 'pool3'

df = load_df(LOGDIR+JSON_FILE)
````

````python
if __name__ == "__main__":
  name = 'data1'
  
  params = Hyperparameters()
  params.num_epochs = 100
  params.hidden_layers = [1024]
  params.initial_learning_rate = 1e-3
  params.cut_layer = CUT_LAYER
  params.check_points_path = CPP+name

  extrc = Extract()
  extrc.return_metrics = True
  extrc.return_losses = True
  
  time_init = time.time()
  output = train(params, extrc)   
  time_conc = time.time()
````

````python
  plt.plot(range(len(output['loss[train]'])), output['loss[train]'], label='train')
  plt.plot(range(len(output['loss[validation]'])), output['loss[validation]'], label='validation')
  plt.title('LOSSES')
  plt.legend()
  plt.savefig(LOGDIR+'losses_normal.png')
  plt.show()
````

````python
  save_params(name, params, output, elapsed_time=time_conc, df=df, path=LOGDIR+JSON_FILE)
````
